```python
# Étape 1 : Extraire le mois et l'année depuis 'Night'
data['Month'] = data['Night'].dt.to_period('M')  # Format YYYY-MM pour grouper par mois

# Étape 2 : Filtrer uniquement les réservations CheckedOut
checked_out = data[data['status'] == 'CheckedOut']

# Étape 3 : Calculer les nuits vendues (nights_sold) par hôtel et par mois
nights_sold = checked_out.groupby(['Hotel', 'Month'])['Night'].count()

# Étape 4 : Calculer le revenu total (total_revenue) par hôtel et par mois
total_revenue = checked_out.groupby(['Hotel', 'Month'])['night_cost'].sum()

# Étape 5 : Ajouter la capacité totale (total_capacity)
# Définir le nombre de chambres pour chaque hôtel
hotel_room_counts = {'HKE': 24, 'HKF': 35, 'HSM': 25}

# Nombre de jours par mois
days_in_month = checked_out['Month'].dt.days_in_month

# Calculer la capacité totale
checked_out['total_capacity'] = checked_out['Hotel'].map(hotel_room_counts) * days_in_month

# Grouper la capacité totale par hôtel et mois
total_capacity = checked_out.groupby(['Hotel', 'Month'])['total_capacity'].max()

# Étape 6 : Calculer le taux d'occupation (occupancy_rate)
occupancy_rate = (nights_sold / total_capacity) * 100

# Créer le tableau final
summary_table = pd.DataFrame({
    'nights_sold': nights_sold,
    'total_revenue': total_revenue,
    'total_capacity': total_capacity,
    'occupancy_rate': occupancy_rate
}).reset_index()

# Afficher le tableau
summary_table
```

```python
# Étape 7 : Calculer l'ADR (Average Daily Rate)
summary_table['ADR'] = summary_table['total_revenue'] / summary_table['nights_sold']

# Afficher le tableau avec l'ADR ajouté
summary_table
```

```python
# Étape 8 : Calculer le PEVPAR
summary_table['PEVPAR'] = (summary_table['occupancy_rate'] * summary_table['ADR']) / 100

# Afficher le tableau avec le PEVPAR ajouté
summary_table
```

```python
# Calculer la somme des métriques
total_nights_sold = summary_table['nights_sold'].sum()
total_revenue = summary_table['total_revenue'].sum()
total_capacity = summary_table['total_capacity'].sum()
total_occupancy_rate = summary_table['occupancy_rate'].mean()  # Moyenne du taux d'occupation
total_adr = summary_table['ADR'].mean()  # Moyenne de l'ADR
total_pevpar = summary_table['PEVPAR'].mean()  # Moyenne du PEVPAR

# Afficher les KPIs sous forme de texte dans un Jupyter Notebook
print("Total Nights Sold:", total_nights_sold)
print("Total Revenue:", total_revenue)
print("Total Capacity:", total_capacity)
print("Average Occupancy Rate:", total_occupancy_rate)
print("Average ADR:", total_adr)
print("Average PEVPAR:", total_pevpar)
```

```python
import pandas as pd

# Assumer que le DataFrame est déjà chargé dans 'data'
# Assurez-vous que la colonne 'Night' est au format datetime
data['Night'] = pd.to_datetime(data['Night'])

# Créer une colonne 'Month' au format 'YYYY-MM' pour grouper par mois
data['Month'] = data['Night'].dt.to_period('M')

# Filtrer uniquement les réservations "CheckedOut"
checked_out_data = data[data['status'] == 'CheckedOut']

# Définir le nombre de chambres pour chaque hôtel
hotel_room_counts = {'HKE': 24, 'HKF': 35, 'HSM': 25}

# Calculer le revenu total par hôtel et mois
total_revenue = checked_out_data.groupby(['Hotel', 'Month'])['night_cost'].sum().reset_index(name='total_revenue')

# Calculer le nombre de nuits vendues par hôtel et mois
nights_sold = checked_out_data.groupby(['Hotel', 'Month'])['Night'].count().reset_index(name='nights_sold')

# Calculer la capacité totale par hôtel et mois
# Ajouter une colonne pour le nombre de jours dans le mois
checked_out_data['days_in_month'] = checked_out_data['Night'].dt.days_in_month
checked_out_data['total_capacity'] = checked_out_data['Hotel'].map(hotel_room_counts) * checked_out_data['days_in_month']

# Calculer la capacité totale par hôtel et mois
total_capacity = checked_out_data.groupby(['Hotel', 'Month'])['total_capacity'].max().reset_index(name='total_capacity')

# Calculer l'ADR par hôtel et mois
adr = total_revenue.merge(nights_sold, on=['Hotel', 'Month'])
adr['adr'] = adr['total_revenue'] / adr['nights_sold']

# Vérification avant la fusion
print("ADR DataFrame:")
print(adr.head())

# Calculer le RevPAR par hôtel et mois
revpar = total_revenue.merge(total_capacity, on=['Hotel', 'Month'])
revpar['revpar'] = revpar['total_revenue'] / revpar['total_capacity']

# Vérification avant la fusion
print("RevPAR DataFrame:")
print(revpar.head())

# Fusionner les résultats pour obtenir une table complète
final_result = adr.merge(revpar, on=['Hotel', 'Month'])

# Vérification du résultat final
print("Final Result before selecting columns:")
print(final_result.head())

# Sélectionner les colonnes pertinentes
final_result = final_result[['Hotel', 'Month', 'total_revenue', 'adr', 'revpar']]

# Afficher le résultat final avec ADR, RevPAR, et Total Revenue par mois et hôtel

final_result
```
